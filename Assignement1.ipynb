Ossia plottiamo la mappa dei pickup e del dropoff

```python
conversion_describe_varaible = np.array(data.describe())

BBox_pickup =(conversion_describe_varaible[3,-3], conversion_describe_varaible[7, -3],
              conversion_describe_varaible[3, -4], conversion_describe_varaible[7, -4])

second_variable_conversion = np.array(data)
pickup_lat = second_variable_conversion[:,-6]
pickup_long = second_variable_conversion[:, -5]

dropoff_lat = second_variable_conversion[:, -3]
dropoff_long = second_variable_conversion[:, -2]

mappina = plt.imread("chicago.png")

fig, ax = plt.subplots(figsize = (20,10))

ax.scatter(pickup_long, pickup_lat, zorder =1, alpha =0.2, c ='b', s=10)
#ax.scatter(dropoff_long, dropoff_lat, zorder =1, alpha=0.2, c ='r',s =10)
ax.set_title('Plotting spacial pickup data on Chiacago map')
ax.set_xlim(BBox_pickup[0], BBox_pickup[1])
ax.set_ylim(BBox_pickup[2], BBox_pickup[3])

ax.imshow(mappina, zorder = 0, extent = BBox_pickup, aspect ='equal')
```
