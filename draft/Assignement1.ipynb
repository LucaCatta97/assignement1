```python
BBox_pickup =(data.describe()['Pickup Centroid Longitude'].values[3], data.describe()['Pickup Centroid Longitude'].values[-1],
             data.describe()['Pickup Centroid Latitude'].values[3], data.describe()['Pickup Centroid Latitude'].values[-1])

pickup_lat = data['Pickup Centroid Latitude'].values

pickup_long = data['Pickup Centroid Latitude'].values

dropoff_lat = data['Dropoff Centroid Latitude'].values
dropoff_long = data['Dropoff Centroid Longitude'].values

mappina = plt.imread("chicago.png")

fig, ax = plt.subplots(figsize = (20,10))

ax.scatter(pickup_long, pickup_lat, zorder =1, alpha =0.2, c ='b', s=10)
#ax.scatter(dropoff_long, dropoff_lat, zorder =1, alpha=0.2, c ='r',s =10)
ax.set_title('Plotting spacial pickup data on Chiacago map')
ax.set_xlim(BBox_pickup[0], BBox_pickup[1])
ax.set_ylim(BBox_pickup[2], BBox_pickup[3])

ax.imshow(mappina, zorder = 0, extent = BBox_pickup, aspect ='equal')
```
